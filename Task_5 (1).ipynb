```python
#import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
data = pd.read_csv('/content/US_Accidents_Dec21_Cleaned.csv')
```

```python
#data preprocessing
# Check for missing values
missing_values = data.isnull().sum()

# Remove duplicates
data.drop_duplicates(inplace=True)


# Check data types
data.info()
```

```python
# Descriptive statistics
summary_stats = data.describe()

# Visualize accident distribution by Weather conditions
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_Condition', data=data)
plt.xticks(rotation=45)
plt.xlabel('Weather Type')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Weather Type')
plt.show()

# Calculate correlations
# Note: Correlations are more suitable for numerical data, so we'll skip this for categorical fields.

# Visualize accident distribution by Weather_Timestamp
plt.figure(figsize=(12, 6))
sns.countplot(x='Weather_Timestamp', data=data)
plt.xticks(rotation=45)
plt.xlabel('Weather_Timestamps')
plt.ylabel('Number of Accidents')
plt.title('Accidents by WeatherTimestamp')
plt.show()

# Visualize accident distribution by a Traffic_Signal
plt.figure(figsize=(12, 6))
sns.countplot(x='Traffic_Signal', data=data)
plt.xticks(rotation=45)
plt.xlabel('Traffic_Signal')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Traffic_Signal')
plt.show()
```

```python
#contributing factors analysis
plt.figure(figsize=(12, 6))
sns.countplot(x='Severity', data=data)
plt.xticks(rotation=90)
plt.xlabel('Cause of Accident')
plt.ylabel('Number of Accidents')
plt.title('Accidents by Contributing Factors')
plt.show()

```
